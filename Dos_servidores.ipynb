```python

from random import expovariate
import math


T0 = 100               # tiempo de simulación
l_l = 2                # tasa de llegada (λ)
mu1 = 1.5              # tasa de servicio en S1 (μ1)
mu2 = 1.2              # tasa de servicio en S2 (μ2)

t = 0
n1 = 0                # número en sistema 1 (cola + servicio)
n2 = 0                # número en sistema 2
NLL = 0               # número de llegadas

tll = expovariate(l_l)   # primera llegada
t2 = math.inf
t3 = math.inf

while t < T0:

    # siguiente evento
    t = min(tll, t2, t3)


    # CASO 1: LLEGADA (t == t1)

    if t == tll:
        NLL += 1
        n1 += 1
        print(f"{t:7.3f}  Llega cliente {NLL} → n1={n1}, n2={n2}")

        # Si S1 estaba vacío, inicia servicio
        if n1 == 1:
            Y1 = expovariate(mu1)
            t2 = t + Y1

        # siguiente llegada
        tll = t + expovariate(l_l)


    # CASO 2: FIN DEL SERVICIO EN S1 (t == t2)
    # cliente pasa a S2

    elif t == t2:
        n1 -= 1
        n2 += 1
        print(f"{t:7.3f}  Termina S1 → entra a S2  | n1={n1}, n2={n2}")

        # Si quedan clientes en S1, atender siguiente
        if n1 > 0:
            Y1 = expovariate(mu1)
            t2 = t + Y1
        else:
            t2 = math.inf

        # Si S2 estaba vacío, iniciar servicio
        if n2 == 1:
            Y2 = expovariate(mu2)
            t3 = t + Y2


    # CASO 3: FIN DEL SERVICIO EN S2 (t == t3)
    # cliente sale del sistema

    else:   # t == t3
        n2 -= 1
        print(f"{t:7.3f}  Termina S2 → SALE del sistema | n1={n1}, n2={n2}")

        # Si quedan clientes en S2, atender siguiente
        if n2 > 0:
            Y2 = expovariate(mu2)
            t3 = t + Y2
        else:
            t3 = math.inf
```
