```python
from random import expovariate
import math

# Parámetros
T0 = 10000              # tiempo de simulación
lmb = 2                 # tasa de llegada λ
mu = 3                  # tasa de servicio μ

# Variables de estado
t = 0
n = 0                   # número de clientes en el sistema
NLL = 0                 # número de llegadas

# Eventos iniciales
tll = expovariate(lmb)  # próxima llegada
t_depart = math.inf     # no hay salida inicial

# Para estimar probabilidades
max_states = 20                     # límite para almacenar estados
time_in_state = [0]*max_states      # tiempo acumulado en cada estado
t_prev = 0                          # tiempo del evento anterior

while t < T0:

    # Siguiente evento (llegada o salida)
    t_next = min(tll, t_depart)
    t = t_next

    # Acumular tiempo en el estado actual antes de cambiar
    if n < max_states:
        time_in_state[n] += t - t_prev
    t_prev = t

    #  CASO 1: LLEGADA
    if t == tll:
        NLL += 1
        n += 1
        print(f"{t:7.3f} Llega cliente {NLL} → n={n}")

        # Si el servidor estaba libre, programar salida
        if n == 1:
            Y = expovariate(mu)
            t_depart = t + Y

        # siguiente llegada
        tll = t + expovariate(lmb)

    #  CASO 2: SALIDA
    else:  # t == t_depart
        n -= 1
        print(f"{t:7.3f} Sale cliente → n={n}")

        # Si quedan clientes, iniciar otro servicio
        if n > 0:
            Y = expovariate(mu)
            t_depart = t + Y
        else:
            t_depart = math.inf

# ---------- RESULTADOS FINALES ----------

print("\nProbabilidades estimadas P(n=k):\n")

total_time = sum(time_in_state)

for k, tk in enumerate(time_in_state):
    if tk > 0:
        print(f"P(n={k}) ≈ {tk/total_time:.4f}")
```
